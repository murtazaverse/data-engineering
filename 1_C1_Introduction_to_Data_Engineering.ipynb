# Data Engineering Introduction

Data Engineering is by far the most important section for a complete data journey. Many people pursue data science and data analytics degrees and certifications (I was one of them lol) without knowing where we will get the data from.

## Definition

As the book "Fundamentals of Data Engineering" defines:

"Data engineering is the development, implementation, and maintenance of systems and processes that take in raw data and produce high-quality, consistent information that supports downstream use cases, such as analysis and machine learning. Data engineering is the intersection of security, data management, DataOps, data architecture, orchestration, and software engineering. A data engineer manages the data engineering lifecycle, beginning with getting data from source systems and ending with serving data for use cases, such as analysis or machine learning."

The following are the tools and concepts required for Data Engineering.

### 1. Databases:
- Used to hold large amounts of data.
- Support applications.
- Other databases are used for analysis.
- Tools: MySQL, PostgreSQL.

### 2. Processing:
- Cleans data.
- Aggregates data.
- Joins data.
- Tools: Apache Spark, Hive.

### 3. Scheduling:
- Plans jobs with specific intervals.
- Resolves dependency requirements of jobs.
- Tools: Apache Airflow, OOZIE
